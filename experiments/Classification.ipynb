```python
models = [("svm", "svmSMtemp"), ("knn", "knnSMtemp"), ("randomforest", "rfSMtemp")]
results = []
for model in models:
    dirs = ["data/filtered/doublesplit/exhale_test/", "data/filtered/doublesplit/inhale_test/"]
    correct = 0
    total = 0
    for filename in os.listdir(dirs[0]):
        if aT.file_classification(dirs[0] + filename, model[1],model[0])[0] == 0.0:
            correct += 1
        total += 1
        
    for filename in os.listdir(dirs[1]):
        if aT.file_classification(dirs[1] + filename, model[1],model[0])[0] == 1.0:
            correct += 1
        total += 1
    results.append((model[0], correct/total))
print(results)
```

```python

[Fs, x] = aIO.read_audio_file("New Recording 6.wav")
segments = aS.silence_removal(x, Fs, 0.020, 0.020, smooth_window = 1, weight = 0.3, plot = True)
segments
```

```python
oldAudio = AudioSegment.from_wav("New Recording 6.wav")
ct = 1
for segment in segments:
    newAudio = oldAudio[segment[0]*1000:segment[1]*1000]
    newAudio.export('test'+str(ct)+'.wav', format="wav")
    ct += 1
```
